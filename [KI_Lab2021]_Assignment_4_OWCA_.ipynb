```
import pandas as pd
import torch
import spacy
spacy.prefer_gpu()
nlp_ger = spacy.load("de")
from tqdm import tqdm
import string
import torch
import torch.nn as nn
import random
import time
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from tabulate import tabulate

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```
df = pd.read_csv('https://raw.githubusercontent.com/tblock/10kGNAD/master/articles.csv', names=['temp','Data'], error_bad_lines=False, sep=';')
df = df.drop(columns=['temp'])
df
```

```
nouns = []
for index, row in tqdm(df.iterrows(), total=df.shape[0], position=0, leave=True):
    text = nlp_ger(row.Data)
    for word in text:
      if word.pos_ == 'NOUN':
        nouns.append(word.text)
```

```
print(len(nouns))
nouns = list(dict.fromkeys(nouns))
print(f"Number of German Nouns {len(nouns)}")
all_letters = string.ascii_letters + 'ÄÖÜäöüß'
nouns = [word for word in nouns if all(ch in list(all_letters) for ch in word)] 
print(f"Number of German sanitized Nouns {len(nouns)}")
n_letters = len(all_letters) + 1
```

```
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RNN, self).__init__()
        self.hidden_size = hidden_size
        self.i2h = nn.Linear(input_size + hidden_size, hidden_size)
        self.i2o = nn.Linear(input_size + hidden_size, output_size)
        self.o2o = nn.Linear(hidden_size + output_size, output_size)
        self.dropout = nn.Dropout(0.1)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, input, hidden):
        input_combined = torch.cat((input, hidden), 1)
        hidden = self.i2h(input_combined)
        output = self.i2o(input_combined)
        output_combined = torch.cat((hidden, output), 1)
        output = self.o2o(output_combined)
        output = self.dropout(output)
        output = self.softmax(output)
        return output, hidden

    def initHidden(self):
        return torch.zeros(1, self.hidden_size)
```

```
# One-hot matrix of first to last letters (not including EOS) for input
def inputTensor(line):
    tensor = torch.zeros(len(line), 1, n_letters)
    for li in range(len(line)):
        letter = line[li]
        tensor[li][0][all_letters.find(letter)] = 1
    return tensor

# LongTensor of second letter to end (EOS) for target
def targetTensor(line):
    letter_indexes = [all_letters.find(line[li]) for li in range(1, len(line))]
    letter_indexes.append(n_letters - 1) # EOS
    return torch.LongTensor(letter_indexes)

def randomTrainingPair():
    line = randomChoice(nouns)
    return line
    
def randomChoice(l):
    return l[random.randint(0, len(l) - 1)]
```

```
# Make category, input, and target tensors from a random category, line pair
def randomTrainingExample():
    line = randomTrainingPair()
    input_line_tensor = inputTensor(line)
    target_line_tensor = targetTensor(line)
    
    input_line_tensor = input_line_tensor.cuda()
    target_line_tensor = target_line_tensor.cuda()
    return input_line_tensor, target_line_tensor
```

```
input, output = randomTrainingExample()
print(input.shape)
```

```
criterion = nn.NLLLoss()

learning_rate = 0.0005

def train(input_line_tensor, target_line_tensor):
    target_line_tensor.unsqueeze_(-1)
    hidden = rnn.initHidden()

    rnn.zero_grad()

    loss = 0

    for i in range(input_line_tensor.size(0)):
        output, hidden = rnn(input_line_tensor[i], hidden)
        l = criterion(output, target_line_tensor[i])
        loss += l

    loss.backward()

    for p in rnn.parameters():
        p.data.add_(p.grad.data, alpha=-learning_rate)

    return output, loss.item() / input_line_tensor.size(0)
```

```
def timeSince(since):
    now = time.time()
    s = now - since
    m = math.floor(s / 60)
    s -= m * 60
    return '%dm %ds' % (m, s)
```

```
rnn = RNN(n_letters, 128, n_letters)

n_iters = 100000
print_every = 10000
plot_every = 500
all_losses = []
total_loss = 0 # Reset every plot_every iters

start = time.time()
for iter in tqdm(range(1, n_iters + 1),position=0, leave=True):
    output, loss = train(*randomTrainingExample()) 
    total_loss += loss

    if iter % print_every == 0:
        print('%s (%d %d%%) %.4f' % (timeSince(start), iter, iter / n_iters * 100, loss))

    if iter % plot_every == 0:
        all_losses.append(total_loss / plot_every)
        total_loss = 0
```

```
max_length = 20

# Sample from a category and starting letter
def sample(start_letters='A'):
    with torch.no_grad():  # no need to track history in sampling
        hidden = rnn.initHidden()
        
        # set output to current prefix
        generated_word = start_letters
        for letter in start_letters[:-1]:
          input_single_letter = inputTensor(letter)
          _ , hidden = rnn(input_single_letter[0], hidden)

        input = inputTensor(start_letters[-1])
        for i in range(max_length):
            output, hidden = rnn(input[0], hidden)
            topv, topi = output.topk(1)
            topi = topi[0][0]
            if topi == n_letters - 1:
              #reached end of line
                break
            else:
                letter = all_letters[topi]
                generated_word += letter
            #next input
            input = inputTensor(letter)

        return generated_word
```

```
from torch.distributions.categorical import Categorical

letter_list = list(all_letters)
letter_list.append("EOS")
import numpy as np
# Sample from a category and starting letter
def sample_percentage(start_letters='A'):
    with torch.no_grad():  # no need to track history in sampling
        hidden = rnn.initHidden()
        
        # set output to current prefix
        generated_word = start_letters

        for letter in start_letters[:-1]:
          input_single_letter = inputTensor(letter)
          _ , hidden = rnn(input_single_letter[0], hidden)

        input = inputTensor(start_letters[.1])
        for i in range(max_length):
            output, hidden = rnn(input[0], hidden)
            m = Categorical(torch.exp(output))
            char = letter_list[m.sample().cpu().numpy()[0]]
            if char == 'EOS':
                break
            input = inputTensor(char)
            generated_word += char

        return generated_word
```

```
def evaluate_model(wordcount = 100, min_prefix =3):
  data = []
  for i in tqdm(range(wordcount)):
    test_word = randomChoice(nouns)
    if len(test_word)>=3:
      prefix_len = random.randint(2, len(test_word)-1)
      test_prefix = test_word[:prefix_len]
      data.append([test_word, test_prefix, test_prefix+"".join((random.choices(all_letters, k=7))), sample(test_prefix) , sample_percentage(test_prefix)])
    else:
      i -=1
  print(tabulate(data, headers=['Word', 'Prefix','Random', "Top", "Distribution"], tablefmt='orgtbl'))

```

```
test_data_top = []
test_data_distribution = []

for i in tqdm(range(10000)):
  test_word = randomChoice(nouns)
  prefix_len = random.randint(1, len(test_word))
  test_prefix = test_word[:prefix_len]
  word_1 = sample(test_prefix)
  word_2 = sample_percentage(test_prefix)

  text = nlp_ger(word_1)
  for word in text:
    if word.pos_ == 'NOUN':
      test_data_top.append(True)
  text = nlp_ger(word_2)
  for word in text:
    if word.pos_ == 'NOUN':
      test_data_distribution.append(True)


```
